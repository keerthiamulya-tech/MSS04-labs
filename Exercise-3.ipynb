```python
def l_term(X, Y, Z):

    return (X and not Y) or (not (Z or Y) or X)
    
print(l_term(True, True, False))
print(l_term(False, True, False))
print(l_term(False, False, True))
```

```python
from sympy import symbols

X, Y, Z = symbols('X Y Z')

expr = (X & ~Y) | (~(Z | Y) | X)

print("Expression:")
print(expr)

print("\nUsing True / False:")

print("(True, True, False) ->",
      expr.subs({X:True, Y:True, Z:False}))

print("(False, True, False) ->",
      expr.subs({X:False, Y:True, Z:False}))

print("(False, False, True) ->",
      expr.subs({X:False, Y:False, Z:True}))

print("\nUsing 1 / 0:")

print("(1,1,0) ->",
      expr.subs({X:1, Y:1, Z:0}))

print("(0,1,0) ->",
      expr.subs({X:0, Y:1, Z:0}))

print("(0,0,1) ->",
      expr.subs({X:0, Y:0, Z:1}))
```

```python
from sympy import *
from sympy.logic.boolalg import *

A, B = symbols('A B')

expr1 = (A | ~B) & ~(~A & B)

print("Expression:")
print(expr1)

print("\nTruth Table:")

for row in truth_table(expr1,[A,B]):
    print(row)

print("\nSimplified:")
simp1 = simplify_logic(expr1)

print(simp1)

print("\nTruth Table Simplified:")

for row in truth_table(simp1,[A,B]):
    print(row)
```
