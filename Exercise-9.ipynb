```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-2, 2, 101)

y = sigmoid(x)

plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.title("Sigmoid")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid()

plt.show()
```
