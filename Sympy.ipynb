```python
def l_term(X, Y, Z):

    return (X and not Y) or (not (Z or Y) or X)
    
print(l_term(True, True, False))
print(l_term(False, True, False))
print(l_term(False, False, True))
```

```python
from sympy import symbols
# symbolic variables
X, Y, Z = symbols('X Y Z')


# boolean expression
expr = (X & ~Y) | (~(Z | Y) | X)

print("Expression:")
print(expr)


# =========================
# Substitution using True/False
# =========================

print("\nUsing True / False:")

print(expr.subs({
    X: True,
    Y: True,
    Z: False
}))

print(expr.subs({
    X: False,
    Y: True,
    Z: False
}))

print(expr.subs({
    X: False,
    Y: False,
    Z: True
}))


# =========================
# Substitution using 1/0
# =========================

print("\nUsing 1 / 0:")

print(expr.subs({
    X: 1,
    Y: 1,
    Z: 0
}))

print(expr.subs({
    X: 0,
    Y: 1,
    Z: 0
}))

print(expr.subs({
    X: 0,
    Y: 0,
    Z: 1
}))
```
