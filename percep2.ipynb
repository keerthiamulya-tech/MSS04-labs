```python
def H(z):

    return 1 if z >= 0 else 0


def output(x1, x2, w1, w2, b):

    z = w1*x1 + w2*x2 + b

    return H(z)


def p_learn(X, Y, w1, w2, b, eta):

    errors = 1

    while errors != 0:

        errors = 0

        for i in range(len(X)):

            x1 = X[i][0]
            x2 = X[i][1]

            y = Y[i]

            a = output(x1, x2, w1, w2, b)

            e = a - y

            if e != 0:

                errors += 1

            # update rule
            if y == 1 and a == 0:

                w1 = w1 + eta * x1
                w2 = w2 + eta * x2
                b = b + eta

            if y == 0 and a == 1:

                w1 = w1 - eta * x1
                w2 = w2 - eta * x2
                b = b - eta

    return w1, w2, b


# OR dataset
X = [[1,1],[1,0],[0,1],[0,0]]

Y_OR = [1,1,1,0]

w1, w2, b = p_learn(X, Y_OR, 0, 0, 0, 0.1)

print("Weights:", w1, w2)
print("Bias:", b)
```

```python
import numpy as np
import matplotlib.pyplot as plt


def H(z):

    return 1 if z >= 0 else 0


def output(x, w, b):

    z = np.dot(x, w) + b

    return H(z)


def p_learn(X, Y, w, b, eta=0.1, epochs=10):

    err_list = []

    for epoch in range(epochs):

        errors = 0

        for x, y in zip(X, Y):

            a = output(x, w, b)

            e = a - y

            if e != 0:

                errors += 1

            # update rule
            if y == 1 and a == 0:

                w = w + eta * x
                b = b + eta

            if y == 0 and a == 1:

                w = w - eta * x
                b = b - eta

        err_list.append(errors)

    return w, b, err_list


def plot_boundary(X, Y, w, b):

    x_min, x_max = -1, 2
    y_min, y_max = -1, 2

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 100),
        np.linspace(y_min, y_max, 100)
    )

    Z = w[0]*xx + w[1]*yy + b

    Z = np.where(Z >= 0, 1, 0)

    plt.contourf(xx, yy, Z, alpha=0.3)

    for x, y in zip(X, Y):

        if y == 1:
            plt.scatter(x[0], x[1], marker='o')

        else:
            plt.scatter(x[0], x[1], marker='x')

    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Perceptron Decision Boundary")

    plt.show()


# OR dataset
X = np.array([
    [1,1],
    [1,0],
    [0,1],
    [0,0]
])

Y_OR = np.array([1,1,1,0])

w = np.zeros(2)

b = 0

w, b, err = p_learn(X, Y_OR, w, b)

print("Weights:", w)
print("Bias:", b)
print("Errors:", err)

plot_boundary(X, Y_OR, w, b)
```

```python
import numpy as np
import matplotlib.pyplot as plt


class perceptron:

    def __init__(self, n_inputs):

        # number of inputs
        self.n_inputs = n_inputs

        # initialize weights with zeros
        self.weights = np.zeros(n_inputs)

        # initialize bias
        self.bias = 0


    def out(self, X):

        # compute weighted sum
        z = np.dot(X, self.weights) + self.bias

        # apply heaviside function
        return 1 if z >= 0 else 0


    def learn(self, data_learn, eta=0.1, epochs=10):

        err_list = []

        for epoch in range(epochs):

            errors = 0

            for x, y in data_learn:

                # perceptron output
                a = self.out(x)

                # compute error
                e = a - y

                if e != 0:

                    errors += 1

                # update weights and bias
                if y == 1 and a == 0:

                    self.weights = self.weights + eta * x
                    self.bias = self.bias + eta

                if y == 0 and a == 1:

                    self.weights = self.weights - eta * x
                    self.bias = self.bias - eta

            err_list.append(errors)

        return err_list


def plot_boundary(X, Y, p):

    x_min, x_max = -1, 2
    y_min, y_max = -1, 2

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 100),
        np.linspace(y_min, y_max, 100)
    )

    Z = p.weights[0] * xx + p.weights[1] * yy + p.bias

    Z = np.where(Z >= 0, 1, 0)

    plt.contourf(xx, yy, Z, alpha=0.3)

    for x, y in zip(X, Y):

        if y == 1:
            plt.scatter(x[0], x[1], marker='o')

        else:
            plt.scatter(x[0], x[1], marker='x')

    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Perceptron Decision Boundary")

    plt.show()


# Input dataset
X = np.array([
    [1,1],
    [1,0],
    [0,1],
    [0,0]
])


# OR dataset
Y_OR = np.array([1,1,1,0])

# AND dataset
Y_AND = np.array([1,0,0,0])

# NAND dataset
Y_NAND = np.array([0,1,1,1])


# =========================
# OR
# =========================

p_or = perceptron(2)

data_or = list(zip(X, Y_OR))

errors_or = p_or.learn(data_or, eta=0.1, epochs=10)

print("OR Weights:", p_or.weights)
print("OR Bias:", p_or.bias)
print("OR Errors:", errors_or)

plot_boundary(X, Y_OR, p_or)


# =========================
# AND
# =========================

p_and = perceptron(2)

data_and = list(zip(X, Y_AND))

errors_and = p_and.learn(data_and, eta=0.1, epochs=10)

print("AND Weights:", p_and.weights)
print("AND Bias:", p_and.bias)
print("AND Errors:", errors_and)

plot_boundary(X, Y_AND, p_and)


# =========================
# NAND
# =========================

p_nand = perceptron(2)

data_nand = list(zip(X, Y_NAND))

errors_nand = p_nand.learn(data_nand, eta=0.1, epochs=10)

print("NAND Weights:", p_nand.weights)
print("NAND Bias:", p_nand.bias)
print("NAND Errors:", errors_nand)

plot_boundary(X, Y_NAND, p_nand)
```
